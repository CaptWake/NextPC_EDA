```python
import numpy as np
import pandas as pd
import fastplot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('test.csv')
```

```python
df['model'].unique()
```

```python
df['model'] = df['model'].map({'swinv2-base-patch4-window16-256': 'B',
       'swinv2-large-patch4-window12-192-22k': 'L',
       'swinv2-small-patch4-window16-256': 'S',
       'swinv2-tiny-patch4-window16-256': 'T'})
```

```python
max_acc_rows = df.groupby(['model', 'preprocess_mode'])['test_acc'].idxmax()
```

```python
sns.scatterplot(df.loc[max_acc_rows], x='test_rec', y='test_prec', hue='preprocess_mode', style='model')
```

```python
# # choose a palette you like https://seaborn.pydata.org/tutorial/color_palettes.html
# cmap = sns.color_palette("flare", n_colors=df["preprocess_mode"].nunique())

# for hue, color in zip(df["preprocess_mode"].unique(), cmap):
#     temp_df = df[df["preprocess_mode"]==hue]
#     plt.plot(temp_df["test_rec"], temp_df["test_prec"], marker="o", color=color)
```

```python
def my_callback(plt):
    ax = sns.scatterplot(df.loc[max_acc_rows], alpha=0.8, x='test_rec', y='test_prec', hue='preprocess_mode', style='model')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    ax.legend(loc='upper center', 
              bbox_to_anchor=(0.5, -0.22), 
              labelspacing=1, 
              ncol=2, 
              borderpad=1.3,
              handletextpad=0.1, 
              prop={'size': 5})
    
fastplot.plot(None,  'prec_rec.png', mode = 'callback', callback = my_callback, grid=True,
                  style='latex', ylim=(0.42, 0.95), figsize=(3,4))
```

```python
max_acc_rows = df.groupby(['model', 'preprocess_mode'])['test_acc'].idxmax()
df.loc[max_acc_rows]
```

```python
def my_callback(plt):
    ax=sns.barplot(df.loc[max_acc_rows], x='model', y='test_acc', hue='preprocess_mode')
    ax.set_ylabel('Accuracy')
    ax.set_xlabel('Model')
    ax.legend(loc='upper center', 
              bbox_to_anchor=(0.5, -0.15), 
              labelspacing=1, 
              ncol=2, 
              borderpad=1.3,
              handletextpad=0.3, 
              prop={'size': 5})
fastplot.plot(None,  'barplot_model_acc_preprocess_mode.png', mode = 'callback', callback = my_callback, grid=True,
              style='latex', ylim=(0.45,0.9), figsize=(3,4))
```
